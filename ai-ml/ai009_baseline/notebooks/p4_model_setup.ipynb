## Section 1 – Setup: Point Python to the Right Folders

Adds AI009's `src/` to `sys.path` and finds the Phoenix repo root regardless of where this notebook is run from.

```python
import sys
from pathlib import Path

# Automatically finds the Phoenix root regardless of who runs it or where they cloned it
PHOENIX_ROOT = Path().resolve()
while not (PHOENIX_ROOT / "ai-ml").exists():
    PHOENIX_ROOT = PHOENIX_ROOT.parent
    if PHOENIX_ROOT == PHOENIX_ROOT.parent:
        raise FileNotFoundError(
            "Could not find Phoenix root. Make sure you're running this from inside the Phoenix repo."
        )

AI009_ROOT = PHOENIX_ROOT / "ai-ml" / "ai009_baseline"
AI009_SRC = AI009_ROOT / "src"

for path in [AI009_ROOT, AI009_SRC]:
    if str(path) not in sys.path:
        sys.path.insert(0, str(path))

print("=== PATH CHECK ===")
print(f"Phoenix root exists : {PHOENIX_ROOT.exists()}")
print(f"AI009 root exists   : {AI009_ROOT.exists()}")
print(f"AI009 src exists    : {AI009_SRC.exists()}")
```
