```python
import os
import sys
import joblib
import pandas as pd

sys.path.append("../src")

from models.isolation_forest import AI012IsolationForest
```

```python
AI012_ROOT = "/workspaces/Phoenix/ai-ml/models/ai012-anomaly"

DATASET_PATH = os.path.join(
    AI012_ROOT,
    "data",
    "raw",
    "anomaly_detection_hourly_2020_2024.csv"
)

CHECKPOINT_PATH = os.path.join(
    AI012_ROOT,
    "checkpoints",
    "isolation_forest.pkl"
)

OUTPUT_DIR = os.path.join(
    AI012_ROOT,
    "data",
    "outputs"
)

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
df = pd.read_csv(
    DATASET_PATH,
    low_memory=False
)

model = joblib.load(CHECKPOINT_PATH)

test_output = model.build_output(df)

test_output.head()
```

```python
test_predictions_path = os.path.join(
    OUTPUT_DIR,
    "isolation_forest_test_predictions.csv"
)

test_output.to_csv(
    test_predictions_path,
    index=False
)

print("Testing completed.")
print("Rows tested:", len(test_output))
print("Anomalies detected:", int(test_output["anomaly_flag"].sum()))
print("Saved to:", test_predictions_path)
```
